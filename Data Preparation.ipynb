```python
import os
import shutil # for copy and paste 
import glob
from tqdm import tqdm
```

```python
Raw_dir = r'H:\Driver_Drowsiness_Detection\mrlEyes_2018_01'
for dirpath, dirname, filenames in os.walk(Raw_dir):
    for i in tqdm([f for f in filenames if f.endswith('.png')]):
        if i.split('_')[4]=='0':
            shutil.copy(src=dirpath+'/'+i, dst=r'H:\Driver_Drowsiness_Detection\Prepared_data\Close_Eye')
        
        elif i.split('_')[4]=='1':
            shutil.copy(src=dirpath+'/'+i, dst=r'H:\Driver_Drowsiness_Detection\Prepared_data\Open_Eye')
```
